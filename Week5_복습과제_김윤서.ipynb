6. 코드에서 데이터를 다룰 때에 사용한 TensorDataset, SequentialSampler, DataLoader 클래스가 무엇인지
> **TensorDataset**: PyTorch에서 여러 개의 텐서를 하나의 데이터셋으로 묶어주는 클래스이다. 같은 인덱스의 데이터끼리 묶이기 때문에 서로 쌍을 이루는 텐서들이 입력되어야 한다. 코드에서는 TensorDataset(train_inputs, train_masks, train_labels)와 같이 세 텐서를 하나의 데이터셋으로 묶는 데 사용했다.   
> **SequentialSampler**: 데이터셋의 인덱스를 순서대로(0, 1, 2, ...) 샘플링하는 클래스이다. 학습 시에는 매 에폭마다 데이터를 섞어야 과적합을 방지할 수 있으므로 RandomSampler를 사용하지만, 검증/테스트 시에는 데이터 순서가 결과에 영향을 주지 않아야 하므로 순서대로 입력되는 SequentialSampler를 사용한다.   
> **DataLoader**: TensorDataset과 Sampler를 받아서, 실제 학습 루프에서 데이터를 배치 단위로 꺼내올 수 있게 해주는 iterator 역할의 클래스이다. batch_size=32로 설정하면 루프를 돌 때마다 32개씩 데이터를 자동으로 묶어서 반환해준다.

> 토크나이저와 분류를 위한 BERT 모델 모두 bert-base-multilingual-cased 사전학습 모델을 불러와 사용했다. 이 모델은 Wikipedia에서 article이 가장 많은 104개의 언어 데이터를 사용하여 MLM 과제로 사전학습된 bert-base 모델이다. 여기서 cased는 대소문자를 구분하는 민감한 모델임을 의미한다. 선택 가능한 다른 BERT 모델은 다음과 같다.   
> - bert-base-uncased: 영어 데이터/MLM 사전학습/대소문자 구분 없음/bert-base
> - bert-large-uncased: bert-base-uncased와 동일한 설정에서 모델 크기만 키운 bert-large 모델
> - bert-base-chinese: bert-base-uncased와 동일한 설정에서 영어 대신 중국어 데이터로 학습시킨 모델
> - bert-large-cased-whole-word-masking: bert-large-cased 모델과 동일한 설정에서 모든 토큰이 한번씩은 마스킹되는 Whole Word Masking 기법을 적용한 모델


8. 모델 학습 loss를 어떤 방식으로 설정해두었는지
> BertForSequenceClassification 모델에 정답 라벨(labels=b_labels)을 함께 넣어 forward를 수행하면, 모델이 내부적으로 loss를 계산하여 반환하는 방식이다. 라벨 개수에 따라 적합한 loss 함수를 선택해 계산하는데, 해당 과제의 라벨은 2개이므로 CrossEntropyLoss를 사용하여 loss를 계산한다. 반환 방식은 라벨 입력 여부에 따라 서로 다른데, 라벨이 같이 입력되는 train의 경우, outputs[0]이 loss, outputs[1]이 logits에 해당하는 반면, 라벨이 입력되지 않는 validation/test의 경우, outputs[0]이 logits에 해당한다. 따라서 학습 loss는 outputs[0]으로 손쉽게 가져올 수 있다.   
참고: https://github.com/huggingface/transformers/blob/a553395766001116a719c82870171f8d6b458c98/src/transformers/models/bert/modeling_bert.py#L1076


9. 모델 성능지표에는 어떤 것을 사용하고 있고 어떻게 구현하고 있는지
> 성능 지표로 정확도(Accuracy)를 사용하며, flat_accuracy라는 함수로 구현하였다. 구체적인 동작 방식은 다음과 같다. 먼저 np.argmax(preds, axis=1)로 배치 내 각 샘플의 logits 중 가장 큰 값의 인덱스(0 또는 1)를 예측 라벨로 추출한다. 이후 예측 라벨과 정답 라벨을 1차원 배열로 펼친 뒤, 두 배열에서 같은 위치의 값이 일치하는 개수를 전체 샘플 수로 나누어 정확도를 계산한다.
    ```python
    def flat_accuracy(preds, labels):
        pred_flat = np.argmax(preds, axis=1).flatten()
        labels_flat = labels.flatten()
        return np.sum(pred_flat == labels_flat) / len(labels_flat)
    ```


10. 코드 중간에 나오는 model.zero_grad()와 model.eval()이 각각 의미하는 바가 무엇인지
> **model.zero_grad()**: 모든 파라미터에 저장된 gradient를 0으로 초기화하는 역할을 한다. PyTorch에서 역전파를 수행하면 각 파라미터에 그래디언트 값이 누적되는데, 이를 초기화하지 않으면 이전 배치의 그래디언트가 현재 배치의 그래디언트에 더해져 잘못된 방향으로 가중치가 업데이트된다. 따라서 model.zero_grad()로 그래디언트를 0으로 초기화하여 각 배치의 그래디언트가 독립적으로 계산되도록 한다.   
> **model.eval()**: 딥러닝 모델에서 학습할 때만 필요한 Dropout, BatchNorm 같은 특수 레이어들을 비활성화하여 추론을 수행할 수 있도록 평가 모드로 전환하는 역할을 한다.


11. output으로 나오는 'logits'은 어떤 형태로 나오며, 해당 벡터를 구성하는 각 차원의 값이 의미하는 바는 무엇인지
> logits는 (배치 사이즈, 클래스 수) 형태의 2차원 배열로 출력된다. 해당 코드에서는 batch_size=32, num_labels=2이므로 (32, 2) 형태가 된다.
각 행은 하나의 입력 문장에 대한 예측 결과이며, 2개의 열은 각 클래스에 해당하는 점수를 의미한다. 즉 인덱스 0의 값은 부정일 가능성에 대한 점수, 인덱스 1의 값은 긍정일 가능성에 대한 점수이다.


12. 모델을 사용하여 inference(새로운 문장 테스트)하는 과정에서 np.argmax(logits) 코드가 수행하는 역할이 무엇인지
> np.argmax(logits)는 logits 배열에서 가장 큰 값의 인덱스를 반환하여 최종 예측 클래스를 결정한다. 예를 들어 logits가 [[-1.2, 3.5]]라면 np.argmax는 1을 반환하여 긍정으로 분류하고, [[ 2.1, -0.8]]이라면 0을 반환하여 부정으로 분류한다.


13. 추론 과정이 아닌 학습 과정에서 np.argmax(logits)가 사용되지 않고 logits가 직접 loss 계산에 사용되는 이유가 무엇인지
> 학습 과정에서는 모델이 얼마나 틀렸는지, 그 정도를 수치화하고 오차를 역전파하여 가중치를 수정해야 한다. np.argmax(logits)는 최댓값의 위치만 반환하는 이산적인 연산이라서 미분이 불가능하며 크기 비교가 불가능하지만, logits는 실수 값 자체를 사용해 CrossEntropyLoss를 계산하여 미분이 가능하며 크기 차이도 학습에 반영할 수 있다. 따라서 미분을 통한 역전파와 효율적인 학습을 위해 logits를 직접 loss 계산에 사용한다.



14. 우리가 사용한 BERT 모델의 config 내용을 살펴보고, config란 무엇인지, 우리 모델의 config를 구성하는 각 요소들이 의미하는 바가 무엇인지 주석으로 설명을 달아주세요. (참고:  https://huggingface.co/google-bert/bert-base-multilingual-cased/blob/main/config.json)


# **모델 학습**

````
# 정확도 계산 함수
def flat_accuracy(preds, labels):

    pred_flat = np.argmax(preds, axis=1).flatten()
    labels_flat = labels.flatten()

    return np.sum(pred_flat == labels_flat) / len(labels_flat)
````

````
# 시간 표시 함수
def format_time(elapsed):

    # 반올림
    elapsed_rounded = int(round((elapsed)))

    # hh:mm:ss으로 형태 변경
    return str(datetime.timedelta(seconds=elapsed_rounded))
````

````
# 재현을 위해 랜덤시드 고정
seed_val = 42
random.seed(seed_val)
np.random.seed(seed_val)
torch.manual_seed(seed_val)
torch.cuda.manual_seed_all(seed_val)

# 그래디언트 초기화
model.zero_grad()

# 에폭만큼 반복
for epoch_i in range(0, epochs):

    # ========================================
    #               Training
    # ========================================

    print("")
    print('======== Epoch {:} / {:} ========'.format(epoch_i + 1, epochs))
    print('Training...')

    # 시작 시간 설정
    t0 = time.time()

    # 로스 초기화
    total_loss = 0

    # 훈련모드로 변경
    model.train()

    # 데이터로더에서 배치만큼 반복하여 가져옴
    for step, batch in enumerate(train_dataloader):
        # 경과 정보 표시
        if step % 500 == 0 and not step == 0:
            elapsed = format_time(time.time() - t0)
            print('  Batch {:>5,}  of  {:>5,}.    Elapsed: {:}.'.format(step, len(train_dataloader), elapsed))

        # 배치를 GPU에 넣음
        batch = tuple(t.to(device) for t in batch)

        # 배치에서 데이터 추출
        b_input_ids, b_input_mask, b_labels = batch

        # Forward 수행
        outputs = model(b_input_ids,
                        token_type_ids=None,
                        attention_mask=b_input_mask,
                        labels=b_labels)

        # 로스 구함
        loss = outputs[0]

        # 총 로스 계산
        total_loss += loss.item()

        # Backward 수행으로 그래디언트 계산
        loss.backward()

        # 그래디언트 클리핑
        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)

        # 그래디언트를 통해 가중치 파라미터 업데이트
        optimizer.step()

        # 스케줄러로 학습률 감소
        scheduler.step()

        # 그래디언트 초기화
        model.zero_grad()

    # 평균 로스 계산
    avg_train_loss = total_loss / len(train_dataloader)

    print("")
    print("  Average training loss: {0:.2f}".format(avg_train_loss))
    print("  Training epcoh took: {:}".format(format_time(time.time() - t0)))

    # ========================================
    #               Validation
    # ========================================

    print("")
    print("Running Validation...")

    #시작 시간 설정
    t0 = time.time()

    # 평가모드로 변경
    model.eval()

    # 변수 초기화
    eval_loss, eval_accuracy = 0, 0
    nb_eval_steps, nb_eval_examples = 0, 0

    # 데이터로더에서 배치만큼 반복하여 가져옴
    for batch in validation_dataloader:
        # 배치를 GPU에 넣음
        batch = tuple(t.to(device) for t in batch)

        # 배치에서 데이터 추출
        b_input_ids, b_input_mask, b_labels = batch

        # 그래디언트 계산 안함
        with torch.no_grad():
            # Forward 수행
            outputs = model(b_input_ids,
                            token_type_ids=None,
                            attention_mask=b_input_mask)

        # 출력 로짓 구함
        logits = outputs[0]

        # CPU로 데이터 이동
        logits = logits.detach().cpu().numpy()
        label_ids = b_labels.to('cpu').numpy()

        # 출력 로짓과 라벨을 비교하여 정확도 계산
        tmp_eval_accuracy = flat_accuracy(logits, label_ids)
        eval_accuracy += tmp_eval_accuracy
        nb_eval_steps += 1

    print("  Accuracy: {0:.2f}".format(eval_accuracy/nb_eval_steps))
    print("  Validation took: {:}".format(format_time(time.time() - t0)))

print("")
print("Training complete!")
````

# **새로운 문장 테스트**

````
# 입력 데이터 변환
def convert_input_data(sentences):

    # BERT의 토크나이저로 문장을 토큰으로 분리
    tokenized_texts = [tokenizer.tokenize(sent) for sent in sentences]

    # 입력 토큰의 최대 시퀀스 길이
    MAX_LEN = 128

    # 토큰을 숫자 인덱스로 변환
    input_ids = [tokenizer.convert_tokens_to_ids(x) for x in tokenized_texts]

    # 문장을 MAX_LEN 길이에 맞게 자르고, 모자란 부분을 패딩 0으로 채움
    input_ids = pad_sequences(input_ids, maxlen=MAX_LEN, dtype="long", truncating="post", padding="post")

    # 어텐션 마스크 초기화
    attention_masks = []

    # 어텐션 마스크를 패딩이 아니면 1, 패딩이면 0으로 설정
    # 패딩 부분은 BERT 모델에서 어텐션을 수행하지 않아 속도 향상
    for seq in input_ids:
        seq_mask = [float(i>0) for i in seq]
        attention_masks.append(seq_mask)

    # 데이터를 파이토치의 텐서로 변환
    inputs = torch.tensor(input_ids)
    masks = torch.tensor(attention_masks)

    return inputs, masks
````

````
# 문장 테스트
def test_sentences(sentences):

    # 평가모드로 변경
    model.eval()

    # 문장을 입력 데이터로 변환
    inputs, masks = convert_input_data(sentences)

    # 데이터를 GPU에 넣음
    b_input_ids = inputs.to(device)
    b_input_mask = masks.to(device)

    # 그래디언트 계산 안함
    with torch.no_grad():
        # Forward 수행
        outputs = model(b_input_ids,
                        token_type_ids=None,
                        attention_mask=b_input_mask)

    # 출력 로짓 구함
    logits = outputs[0]

    # CPU로 데이터 이동
    logits = logits.detach().cpu().numpy()

    return logits
````
